### Part 1: Building the base Anagram Checker

Given two valid strings, check to see if they are anagrams of each other. If it is, return `True`, else `False`. For this part, we can assume that uppercase letters are the same as if it was a lowercase character.

Examples of anagrams:
* Slient and Listen
* Night and Think

Example outputs:
```python
anagram_checker("Slient", "listen") # True
anagram_checker("Slient", "Night") # False
anagram_checker("night", "Thing") # True
```

````python
# This function checks if the two words passed as parameters are anagrams or not.
# It returns true if word_1 and word_2 are anagrams. It returns false otherwise.
def anagram_checker(word_a, word_b):
  '''
     This function checks if the two words passed as parameters are anagrams or not.
     It returns true if word_a and word_b are anagrams. It returns false otherwise.
     The return value of this function will remain the same regardless of case-sentivity
     of the letters in the word.
     Parameters: 
        word_a: str
        word_b: str
      Returns: 
        bool       
  '''
  lowercase_word_a = word_a.replace(" ", "").lower()    # replace space with nothing
  lowercase_word_b = word_b.replace(" ", "").lower()
  return sorted(lowercase_word_a) == sorted(lowercase_word_b)

anagram_checker("Slient", "listen")

# help(anagram_checker)
````

### Part 2: Expanding the functionality of the Anagram Checker

Using your existing and functional anagram checker, let's add a boolean option called `is_case_sensitive`, which will return `True` or `False` based on if the two compared words are anagrams and if we are checking for case sensitivity.

````python
def anagram_checker(word_a, word_b, is_case_sensitive):
  '''
    This function checks if the two words passed as parameters are anagrams or not.
    It returns true if word_a and word_b are anagrams. It returns false otherwise.
    If the is_case_sensitive argument is True, this function will also make sure that the 
    cases of the letters match. 
    
    E.g: If the words are "Slient" and "listen",
    it will return True if is_case_sensitive is False. It will return False if is_case_sensitive
    is True even though the two words are anagrams. This is because cases of letter 's' do not match
    
    Parameters:
        word_a: str
        word_b: str
        is_case_sensitive: bool (True/False to match the case or not)
    Returns: 
        bool        
  '''
  word_a = word_a.replace(" ", "")
  word_b = word_b.replace(" ", "")

  if is_case_sensitive == True:
    return sorted(word_a) == sorted(word_b)
  else:
    return sorted(word_a.lower()) == sorted(word_b.lower())
    

anagram_checker("Slient", "listen", False) # True
````

````python
anagram_checker("Slient", "Listen", True) # False
#help(anagram_checker)
````

````python
anagram_checker("Dormitory", "dirty room", True) # False
````

````python
anagram_checker("Dormitory", "dirty room", False)  # True
````

````python
anagram_checker("Dormitory", "Dirty room", True)  # True
````

````python
anagram_checker("Dormitory", "Dirty room", False)  # True
````
